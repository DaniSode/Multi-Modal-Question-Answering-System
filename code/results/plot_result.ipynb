## Read and plot loss


Import libraries

```python
import matplotlib.pyplot as plt
```

Read data

```python
train_path = "train_log.txt"
val_path = "val_log.txt"
train_vals = []
val_vals = []
with open(train_path, 'r') as train_data:
    for value in train_data: 
        value = float(value.split('\t')[1].rstrip('\n'))
        train_vals.append(value)

with open(val_path, 'r') as val_data:
    for value in val_data: 
        value = float(value.split('\t')[1].rstrip('\n'))
        val_vals.append(value)
```

Plot result

```python
plt.plot(train_vals)
plt.plot(val_vals)
plt.ylabel('Number of epochs')
plt.xlabel('Loss')
plt.title('Train vs validation loss for Multi-modal Question Answering model')
plt.legend(['Train loss', 'Validation loss'])

plt.savefig("loss_plot.png")
```
